# Slice 5: Model Architecture Verification

This notebook verifies the `DeepfakeDetector` model architecture and performs a dummy forward pass.

```python
import sys
import os
from google.colab import drive
import torch

# 1. Setup Environment
drive.mount('/content/drive')
project_path = '/content/drive/My Drive/deepfakedetection'
sys.path.append(project_path)
os.chdir(project_path)

# Reload imports
%load_ext autoreload
%autoreload 2

from src.model import DeepfakeDetector
```

```python
# 2. Initialize Model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

model = DeepfakeDetector(pretrained=True)
model = model.to(device)

# Print model summary (optional, needs torchsummary but simple print works)
# print(model)
```

```python
# 3. Dummy Forward Pass
# Create random input tensor (Batch Size=4, Channels=3, Height=224, Width=224)
dummy_input = torch.randn(4, 3, 224, 224).to(device)

print(f"Input shape: {dummy_input.shape}")

with torch.no_grad():
    output = model(dummy_input)
    
print(f"Output shape: {output.shape}")
print("Output logits:", output)

assert output.shape == (4, 1), f"Expected output shape (4, 1), got {output.shape}"
print("Model verification successful! ✅")
```
