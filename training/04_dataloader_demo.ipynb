# Slice 4: Dataset & Augmentation Visualization

This notebook verifies that the `DeepfakeDataset` class works correctly and visualizes the augmentations.

```python
import sys
import os
from google.colab import drive
import torch
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup Environment
drive.mount('/content/drive')
project_path = '/content/drive/My Drive/deepfakedetection'
sys.path.append(project_path)
os.chdir(project_path)

# Reload imports
%load_ext autoreload
%autoreload 2

from src.dataset import DeepfakeDataset, get_transforms
```

```python
# 2. Initialize Dataset (Train)
# Using 'data' folder created in Slice 2
data_root = 'data'

train_ds = DeepfakeDataset(
    root_dir=data_root, 
    phase='train', 
    transform=get_transforms(phase='train')
)

print(f"Total samples: {len(train_ds)}")
```

```python
# 3. Create DataLoader
from torch.utils.data import DataLoader

train_loader = DataLoader(train_ds, batch_size=8, shuffle=True, num_workers=2)
images, labels = next(iter(train_loader))

print(f"Batch shape: {images.shape}") # Should be [8, 3, 224, 224]
print(f"Labels: {labels}")
```

```python
# 4. Visualize
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)

# Plot grid
import torchvision

out = torchvision.utils.make_grid(images)
plt.figure(figsize=(12, 12))
class_names = ['Real', 'Fake']
imshow(out, title=[class_names[x] for x in labels])
plt.show()
```
