# Slice 3: Video Processing & Face Extraction Demo

This notebook demonstrates how to use the `src.face_extraction` module to detect and crop faces from videos.

```python
import sys
import os
from google.colab import drive
import matplotlib.pyplot as plt

# 1. Setup Environment
drive.mount('/content/drive')
project_path = '/content/drive/My Drive/deepfakedetection'
sys.path.append(project_path)
os.chdir(project_path)

# Reload imports if changed
%load_ext autoreload
%autoreload 2

from src.face_extraction import FaceExtractor
```

```python
# 2. Initialize Extractor
# This will download MTCNN weights on first run
extractor = FaceExtractor()
```

```python
# 3. Upload a Test Video
from google.colab import files
print("Please upload a short video (mp4) or image (jpg/png) to test face detection.")
uploaded = files.upload()
```

```python
# 4. Process and Visualize
if len(uploaded) > 0:
    filename = list(uploaded.keys())[0]
    print(f"Processing {filename}...")
    
    if filename.lower().endswith(('.mp4', '.avi', '.mov')):
        faces = extractor.process_video(filename, num_frames=5)
        print(f"Found {len(faces)} faces.")
        
        # Plot
        if len(faces) > 0:
            fig, axes = plt.subplots(1, len(faces), figsize=(15, 5))
            if len(faces) == 1:
                axes = [axes]
            for ax, face in zip(axes, faces):
                ax.imshow(face)
                ax.axis('off')
            plt.show()
    else:
        # Assume image
        face = extractor.process_image(filename)
        if face:
            plt.imshow(face)
            plt.axis('off')
            plt.show()
        else:
            print("No face detected.")
else:
    print("No file uploaded.")
```
